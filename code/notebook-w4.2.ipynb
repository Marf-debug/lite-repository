```python
# Map column names to types and descriptions for display
cols_for_display = get_cols_for_display(dataset)

# Interactive widget to select a type of feature and display the corresponding columns
data_types_list = Dropdown(
    options=['Discrete', 'Continuous', 'Boolean', 'Categorical', 'All'],
    description='Select a variable type:',
    disabled=False,
    style={'description_width': 'max-content'},
    value='Discrete'
)

def retrieve_cols_for_display(col_type):
    if col_type != 'All':
        return cols_for_display[(cols_for_display['type'] == col_type)]
    else:
        return cols_for_display

interact(retrieve_cols_for_display, col_type=data_types_list);
```

## Reminder: missing values

Now let's have a look at how much data in our dataset is missing:
* There may be different reasons for why data is missing.
* Sometimes a value can be calcualted from other variables in the dataset.
* Other times a variable may have to be excluded from further analysis altogether.

```python
# Calculate the number of missing values in each column and sort in descending order
missing_values = dataset.isna().sum().sort_values(ascending=False)
missing_values = missing_values[missing_values > 0]
missing_values = missing_values.to_frame(name='Number of missing values')

# Interactive widget to show the number of missing values in each column
button = Button(description ='Show number of missing values', layout=Layout(width='50%'))
button.style.button_color = button_color
output = Output()

def on_button_clicked(_):
    output.clear_output()
    n_values = dataset.shape[0]
    missing_values = dataset.isna().sum().sort_values(ascending=False)
    missing_values = missing_values[missing_values > 0]
    missing_values = missing_values.to_frame(name='Number of missing values')
    missing_values['Proportion of missing values'] = (missing_values['Number of missing values'] / n_values * 100).round(2)
    with output:
        display(missing_values)

button.on_click(on_button_clicked)

display(button)
display(output)
```

## Multivariable analysis

> Refresher:
> 
> Multivariable EDA is used to show relationships between two or more variables, usually the independent variables (features) through a cross-tabulation of statistics.
> A heatmap is an efficient way to visualise correlations between multiple variables.
>
> There are multiple important things to consider when doing correlation analysis:
> * Different statistics are suitable for different types of variables. Here, we will only examine pair-wise relationships between continuous variables. 
> * We will calculate the Pearson's correlation coefficient. The coefficient ranges from -1 to +1, where:
>   - Values closer to -1 mean perfect inverse correlation <=> As one feature increases, the other decreases.
>   - Values closer to 0 mean weak or no correlation between the variables
>   - Values closer to +1 mean perfect positive correlation <=> As one feature increases, the other also increases.
> * It is important to remember that Pearson's correlation coefficient detects only **linear** correlation. If the relationship between two variables is U-shaped, the Pearson's correlation coefficient will likely be close to 0 despite there being a strong non-linear relationship. This is why it is so important to visually examine relationships between variables using scatter plots.
>
> A heatmap helps pinpoint redundant variables. The problem with redundant variables/features when it comes to predictive modelling is that they do not bring new information. Yet, each additional feature increases the complexity of our predictive model and may introduce noise that will ultimately result in a weaker model. 

Let's first look for correlations between indepandent variables (i.e., all variables but the outcome!)

```python
button2 = Button(description ='Create a heatmap to examine correlations between variables', layout=Layout(width='50%'))
button2.style.button_color = button_color
output2 = Output()

cmaps = ['bwr', 'viridis', 'plasma', 'inferno', 'magma', 'Greys', 'Blues','PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu', 'RdYlBu', 'RdYlGn', 'Spectral', 'coolwarm',  'seismic']

cmaps_list = Dropdown(
    options = sorted(cmaps), 
    description='Select a colormap for the heatmap:', 
    disabled=False,
    style={'description_width': 'max-content'}, 
    value = 'bwr')


def on_button_clicked2(_):
    output2.clear_output()
    cols = dataset.select_dtypes(include=['float64']).columns.tolist()
    plt.figure(figsize=(20, 10))
    sns.heatmap(dataset[cols].corr().round(2), annot=True, linewidths=4, cmap=cmaps_list.value)
    plt.xticks(rotation=30, ha="right", rotation_mode="anchor")
    with output2:
        plt.show()

button2.on_click(on_button_clicked2)

display(cmaps_list)
display(button2)
display(output2)
```

## Univariable analysis

> Our ultimate goal is to see if we can safely prevent patient re-admittion to the hospital. 
> * In order to intervene, we would like to know which patient characteristics (variables/features) are linked to the higher chance of being re-admitted. 
> * Ideally, we would like to be able to predict if a patient will be re-admitted, given the information available to us. 
> * Here, we use retrospective data with patient information collected up to 2025 and their re-admission status in 2025 store in `outcome_admission_in_2025`. 
> * We refer to `outcome_admission_in_2025` as our target variable (also called the outcome variable)

Let's see which characteristics/features/variables correlate with the target/outcome variable.
* For categorical and boolean variables, we will plot stacked barplots showing the percentage breakdown across the admitted and non-admitted patients.
* For discrete and continuous variables, we will inspect boxplots comparing distributions across the admitted and non-admitted patients.
* Start with the variables you expect to be correlated with the outcome based on your team's clinical expertise. Then move on to the less likely candidates.
* Does the data show what you expected? *Be ready to share with the rest of the class*

```python
def plot_proportions(col):
    proportions = pd.crosstab(dataset.outcome_admission_in_2025, col, normalize='index') * 100
    ax = proportions.plot(kind='bar', stacked=True, 
                    color=sns.color_palette("Set2"), rot=0, figsize=(4, 5))

    plt.xticks(labels=['Not admitted', 'Admitted'], ticks=[0, 1])
    plt.xlabel('Outcome variable: admission in 2025')
    plt.ylabel('Proportion of patients');
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left');

    for rec in ax.patches:
            height = rec.get_height()
            ax.text(rec.get_x() + rec.get_width() / 2,
                        rec.get_y() + height / 2,
                        "{:.0f}%".format(height),
                        ha='center',
                        va='bottom')

cat_var_list = Dropdown(
     options = sorted(dataset.drop('outcome_admission_in_2025', axis=1).select_dtypes(include=['category', 'bool']).columns.tolist()), 
     description='Select a column to view:', 
     disabled=False,
     style={'description_width': 'max-content'}, 
     value = 'sex'
     )

button3 = Button(description ='Plot relationships between categorical/boolean variables and the outcome', layout=Layout(width='50%'))
button3.style.button_color = button_color
output3 = Output()

def on_button_clicked3(_):
    output3.clear_output()
    plot_proportions(dataset[cat_var_list.value])
    with output3:
        plt.show()

button3.on_click(on_button_clicked3)
display(cat_var_list, button3)
display(output3)
```

```python
def plot_trends(col):
    plt.figure(figsize=(4.1, 5))
    sns.boxplot(x='outcome_admission_in_2025', y=col, data=dataset, palette=sns.color_palette('Set2'))
    plt.xticks(labels=['Not admitted', 'Admitted'], ticks=[0, 1])
    plt.xlabel('Outcome variable: admission in 2025')
    plt.ylabel(col.name.replace('_', ' ').capitalize());
     
num_var_list = Dropdown(
    options=sorted(dataset.drop('patient_id', axis=1).select_dtypes(include=['float64', 'int64']).columns.tolist()),
    description='Select a column to view:',
    disabled=False,
    style={'description_width': 'max-content'},
    value='age'
)

button4 = Button(description ='Plot relationships between discrete/continuous variables and the outcome', layout=Layout(width='50%'))
button4.style.button_color = button_color
output4 = Output()

def on_button_clicked4(_):
    output4.clear_output()
    plot_trends(dataset[num_var_list.value])
    with output4:
            plt.show()

button4.on_click(on_button_clicked4)
display(num_var_list, button4)
display(output4)
```
